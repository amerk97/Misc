```python
import matplotlib.pyplot as plt
import numpy as np


#Yield diameter---------------
yields = [800, 1000, 1200, 1300, 1400, 1600, 1800]

def dy(yieldstrength):
    d1 = ((2*32*3.8*10**6)/(yieldstrength*np.pi))**(1/3)
    #print(str(yieldstrength) + ' Mpa gives dy = ' + str(d1))
    return d1

yieldvals = []
for i in yields:
    yieldvals.append(dy(i))

print('\n')

#Fracture diameter, with variable a:--------------
KICS = [187, 182, 152, 102, 65, 41, 35]

def dk(KIC, a):
    d2 = ((3*32*3.8*10**6*0.728*np.sqrt(np.pi*a))/(KIC*np.pi))**(1/3)
    #print(str(KIC) + ' MPa*sqrt(m) and ' + str(a) + ' m gives dk = ' + str(d2))
    return d2

fracvals = []
fvals05 = []
fvals2 = []
for n in KICS:
    fracvals.append(dk(n, 0.001))
for a in KICS:
    fvals05.append(dk(a, 0.0005))
for e in KICS:
    fvals2.append(dk(e, 0.002)) 
    
    
#plotting 
plt.xlim(600, 2000)
plt.ylim(20, 65)
plt.xlabel('Yield strength MPa')
plt.ylabel('Diameter d, mm')
plt.plot(yields, yieldvals, '-o', label = 'Yield')
plt.plot(yields, fracvals, '-o', label = 'fracture, a=1mm')
plt.plot(yields, fvals05, '-o', label = 'fracture, a=0.5mm')
plt.plot(yields, fvals2, '-o', label = 'fracture, a=2mm')

plt.axvline(x=1277, color='g', linestyle=':')
plt.axhline(y=39.3, color='g', linestyle=':')

plt.grid()
plt.legend()
plt.show()
```
