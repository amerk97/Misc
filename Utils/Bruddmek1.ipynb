```python
import numpy as np
import matplotlib.pyplot as plt

E_0 = 3.2*10**-20
a = 0.38
r = np.linspace(0.3, 1, 10000)
U_r = E_0*((a/r)**12-(a/r)**6)
rep = E_0*(a/r)**12
att = E_0*(-(a/r)**6)

plt.plot(r, U_r, label = 'Potential energy U(r)')
plt.plot(r, rep, label = 'Attractive energy')
plt.plot(r, att, label = 'Repulsive energy')



plt.xlabel('Distance r - [nm]')
plt.ylabel('Energy U(r) - [J]')
plt.legend()
plt.grid()
plt.ylim([-10**-20, 10**-20])
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

E_0 = 3.2*10**-20
a = 0.38
r = np.linspace(0.3, 1, 10000)
F_r = E_0*(-12*a**12*r**-13 + 6*a**6*r**-7)

plt.plot(r, F_r, label = 'Force versus distance r')

plt.xlabel('Distance r - [nm]')
plt.ylabel('Force F(r) - [N]')
plt.legend()
plt.grid()
plt.ylim([-2*10**-20, 5.5*10**-20])
plt.show()
```
